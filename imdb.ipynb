#### Convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP.

```python
movies.replace({'content_rating' : {'NOT RATED':'UNRATED', 'APPROVED':'UNRATED', 'PASSED':'UNRATED','GP':'UNRATED'}}, 
               inplace = True)
movies.head(40)
```

#### Convert the following content ratings to "NC-17": X, TV-MA.

```python
movies.replace({'content_rating' : {'X':'NC17', 'TV-MA':'NC17', 'NC-17':'NC17'}}, inplace = True)
movies.head(40)
```

#### Count the number of missing values in each column.

```python
movies.isnull().sum()
```

#### If there are missing values: examine them, then fill them in with "reasonable" values.

```python
null_movie = movies.loc[:, 'content_rating'].isnull()
movies.loc[null_movie, :]
movies.loc[:, 'content_rating'].fillna(value='PG', inplace=True)
movies.loc[:, 'content_rating'].isnull().sum()
```

#### Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours.

```python
movies.loc[:, 'over_2_hrs'] = (movies.loc[:,'duration'] >= 120)
#movies.drop('Highest Rated', axis=1, inplace=True)
movies.groupby('over_2_hrs').mean()
```
